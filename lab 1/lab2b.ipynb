```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

đọc file csv + xem structure của dataframe + dropna một số cột

```python
a = pd.read_csv('titanic.csv')
a.info()
a.keys()
a = a.dropna(subset=['pclass', 'age', 'sex', 'sibsp', 'survived'])
a.head()
```

biến dự đoán là survived, còn biến data pclass, age, sex và sibsp
lý do chọn các biến độc lập như này là vì khách có sống sót hay không đa số phụ thuộc vào pclass (tier càng cao thì được ưu tiên đi trước, đỡ chen lấn), age (trẻ thì chạy nhanh, khỏe hơn già), sex (nam thì khỏe hơn nữ (đa số trường hợp)), sibsp (không có người nhà thì đỡ bận tâm, lo chạy được, càng nhiều người nhà càng phân vân dễ chìm theo tàu)

```python
x = a[['pclass', 'age', 'sex', 'sibsp']]
y = a['survived']
```

làm model linear regression, chia train/test theo tỉ lệ 80%/20%

```python
model = LinearRegression()
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
model.fit(x_train, y_train)
output = model.predict(x_test)
```

model có độ chính xác khá thấp khi chỉ giải thích được khoảng 36,6% sự biến thiên của dữ liệu và các chỉ số sai số cho thấy hiệu quả dự đoán của mô hình chưa cao.

```python
r2 = r2_score(y_test, output)
mae = mean_absolute_error(y_test, output)
mse = mean_squared_error(y_test, output)
print(r2, mae, mse)
```
